```python
import numpy as np
import tensorflow as tf
```

```python
y_list = []
input_list = []

for W in range(1,30):
    for b in range(30):
        y = [W*x+b for x in range(1,6)]
        y_list.append(y)   
        input_list.append([W,b])

y_list = np.array(y_list)
input_list = np.array(input_list)

X_input = tf.placeholder(shape =[None,2],dtype = tf.float32)
Y_output = tf.placeholder(shape=[None,5],dtype = tf.float32)
```

```python
W1 = tf.get_variable("W1", shape=[2, 64],
                     initializer=tf.contrib.layers.xavier_initializer())
b1 = tf.Variable(tf.random_normal([64]))
L1 = tf.nn.relu(tf.matmul(X_input, W1) + b1)

W2 = tf.get_variable("W2", shape=[64, 64],
                     initializer=tf.contrib.layers.xavier_initializer())
b2 = tf.Variable(tf.random_normal([64]))
L2 = tf.nn.relu(tf.matmul(L1, W2) + b2)

W3 = tf.get_variable("W3", shape=[64, 5],
                     initializer=tf.contrib.layers.xavier_initializer())
b3 = tf.Variable(tf.random_normal([5]))
hypothesis = tf.matmul(L2, W3) + b3


cost = tf.reduce_mean(tf.losses.mean_squared_error(predictions=hypothesis, labels=Y_output))
optimizer = tf.train.AdamOptimizer(learning_rate=0.01).minimize(cost)
```

```python
with tf.Session() as sess:
    
    sess.run(tf.global_variables_initializer())
    
    for epoch in range(1000):
        loss, _ = sess.run([cost,optimizer], feed_dict = 
                           {X_input:input_list, Y_output:y_list})
        
        print("epoch {}, loss {}".format(epoch,loss))
    
    test_input = np.array([[3,1], [1.5,0]])
    test_output = sess.run(hypothesis, feed_dict = {X_input:test_input})
    
    print("#### test ####")
    print("test input1 : {} , test_output1 : {}".format(test_input[0],test_output[0]))
    print("Real test1 output :{}".format([test_input[0][0]*x+test_input[0][1] for x in range(1,6)]))
    print("test input2 : {} , test_output2 : {}".format(test_input[1],test_output[1]))
    print("Real test2 output :{}".format([test_input[1][0]*x+test_input[1][1] for x in range(1,6)]))
```
